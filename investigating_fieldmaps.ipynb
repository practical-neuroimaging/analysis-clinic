```python
dcm_dir = '/Users/mb312/data/gre_field_mapping_4'
dcm_fnames = sorted(os.listdir(dcm_dir))
dcm_fnames
```

```python
dcm_paths = [os.path.join(dcm_dir, fname) for fname in dcm_fnames]
```

```python
from hashlib import sha1
hash_values = {}
for path in dcm_paths:
    contents = open(path, 'rb').read()
    hash_value = sha1(contents).hexdigest()
    basename = os.path.basename(path)
    if hash_value in hash_values:
        print(basename,
              'has same byte contents as',
              hash_values[hash_value])
    else:
        hash_values[hash_value] = basename
```
